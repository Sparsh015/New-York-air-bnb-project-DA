```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("bnb new york dataset.csv", encoding_errors ='ignore')
```

**4. Data  cleaning**

```python
#dropping all missing value rows
data.dropna(inplace=True)

data.isnull().sum()
# data.fillna()
```

```python
#shape after ignoring rows 
data.shape
```

```python
#dealing with duplicates
data.duplicated().sum()

#to show duplicates only
#data[data.duplicated()]

#delete duplicate data
data.drop_duplicates(inplace = True)

data[data.duplicated()]
```

```python
#type casting
# to change the datatype of a column
data['id'] = data['id'].astype(object)

print(data.dtypes)

data['host_id'] = data['host_id'].astype(object)

data.dtypes
```

### EDA
**5. Data Analysis**

**Univariate Analaysis**

```python
#identifying outliers in price

df = data[data['price'] < 1500]
sns.boxplot(data = df, x = 'price')
```

```python
#price distribution
plt.figure(figsize = (8,5))
plt.ylabel('frequency')
plt.title('Price distribution')
sns.histplot(data = df, x = 'price', bins = 100)
plt.show()
```

```python
df.columns
```

```python
#price distribution
plt.figure(figsize = (6,3))
plt.ylabel('frequency')
plt.title('Price distribution')
sns.histplot(data = df, x = 'availability_365')
plt.show()
```

```python
data.dtypes
```

```python
df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending = False)
```

**Feature Engineering**

```python
#price per bed

df['price per bed'] = df['price'] / df['beds']
df.head()
```

```python
df.groupby('neighbourhood_group')['price per bed'].mean().sort_values(ascending = False)
```

**Bi Variabte Analysis**
one variable dependency in another variable 

```python
df.columns
```

```python
#price dependency on neighbourhood
sns.barplot(data=df , x = 'neighbourhood_group',y = 'price',hue = 'room_type')
```

```python
#number of reviews and price relataion
plt.figure(figsize = (8,5))
plt.title("locality and review dependency")
sns.scatterplot(data = df, x = 'number_of_reviews', y = 'price', hue = 'neighbourhood_group')
```
